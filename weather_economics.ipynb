```python
import csv

weather_file_path = "C://Users//vitz2//Desktop//Weather_Data_Jakarta.csv"
file = open(weather_file_path, 'r')
weather = csv.DictReader(file)
```

```python
for row in weather:
    print(row)
```

```python
param = ['datetime','temp', 'precip','cloudcover']
res=[]
for row in weather:
    param_data = dict((key, row[key]) for key in param)
    res.append(param_data)
res
```

```python
import pandas as pd
df = pd.DataFrame(res)

# Convert the 'datetime' column to datetime format
df['datetime'] = pd.to_datetime(df['datetime'])

# Create a new column 'month' extracting the month from the 'datetime' column
df['month'] = df['datetime'].dt.to_period('M')

# Convert 'temp' column to numeric
df['temp'] = pd.to_numeric(df['temp'])
df['precip'] = pd.to_numeric(df['precip'])
df['cloudcover'] = pd.to_numeric(df['cloudcover'])

# Group by 'month' and calculate the average 'temp'
climate_jak = df.groupby('month')['temp','precip','cloudcover'].mean().reset_index()
# Rename the 'month' column to the actual name
climate_jak['month'] = climate_jak['month'].dt.strftime('%Y-%m')
```

```python
eco22_file_path = "C://Users//vitz2//Desktop//Inflation-2022.csv"
file = open(eco22_file_path, 'r',encoding='utf-8-sig')
inflation22 = csv.DictReader(file)
```

```python
for row in inflation22:
    print(row)
```

```python
jak_infl=[]
for row in inflation22:
    if(row['City'] == 'DKI JAKARTA'):
        jak_infl.append(row)
```

```python
eco23_file_path = "C://Users//vitz2//Desktop//Inflation-2023.csv"
file = open(eco23_file_path, 'r',encoding='utf-8-sig')
inflation23 = csv.DictReader(file)
```

```python
for row in inflation23:
    print(row)
```

```python
for row in inflation23:
    if(row['City'] == 'DKI JAKARTA'):
        jak_infl.append(row)
```

```python
df_jak_infl = pd.DataFrame(jak_infl)
df_jak_infl
```

```python
infl_jak=df_jak_infl.transpose()
infl_jak.rename(columns={0:'2022'},inplace=True)
infl_jak.rename(columns={1:'2023'},inplace=True)
infl_jak
```

```python
from datetime import datetime

row=0
for val in climate_jak['month']:
    # Convert the date string to a datetime object
    date_object = datetime.strptime(val, '%Y-%m')
    # Get the month name from the datetime object
    month_name = str(date_object.strftime('%B'))
    if(row<11):
        month_name=month_name+" 2022"
        climate_jak.at[row,'month']=month_name
    else:
        month_name=month_name+" 2023"
        climate_jak.at[row,'month']=month_name
    row=row+1
```

```python
final_data=climate_jak.drop(24)
final_data
```

```python
inflation=[]
ele=0
for val in infl_jak.values:
    if(val[0]!="DKI JAKARTA"):
        inflation.append(val[0])
for val in infl_jak.values:
    if(val[1]!="DKI JAKARTA"):
        inflation.append(val[1])
```

```python
final_data['inflation']=inflation
final_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

X=final_data[['temp', 'precip', 'cloudcover']]
y=final_data['inflation']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=50)
gbm=GradientBoostingRegressor(n_estimators=50,learning_rate=0.02,random_state=50)

gbm.fit(X_train, y_train)
gbm_predict=gbm.predict(X_test)

mse=mean_squared_error(y_test, gbm_predict)
print("Error: ",mse)

print('Feature Importances:')
for feature, importance in zip(X.columns,gbm.feature_importances_):
    print(feature,':',importance)
```

```python
import numpy as np

#Testing for random values of the above parameters to predict the inflation
random_val = np.array([[31.0, 20.0, 70.0]]) 
predict_infl = gbm.predict(random_val)
print('Predicted Inflation: ',predict_infl[0])
```
